```python
import numpy as np
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib as plt
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
data.head() # flipper 지느러미
```

```python
# scatter plot (산점도)
# 숫자들의 관계를 표현하려 함

sns.relplot(data = data, x = 'bill_length_mm', y = 'body_mass_g')
```

```python
sns.relplot(data = data, x = 'bill_length_mm', y = 'body_mass_g', hue = 'species') # hue : 색깔을 다르게 표현하려고 함
```

```python
sns.relplot(data = data, x = 'bill_length_mm', y = 'body_mass_g', hue = 'species', style = 'island') # 모양으로 그룹을 나누기, 잘 구분이 안됨
```

```python
sns.relplot(data = data, x = 'bill_length_mm', y = 'body_mass_g', hue = 'species', col = 'island') # 그림을 3개로 쪼개서 본다
```

```python
sns.relplot(data = data, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'body_mass_g', size = 'body_mass_g', sizes = (10,300))

# 부리가 길고 앏은 애들 - 무게가 많이 나감
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
# 시간에 따라 어떻게 변화하는지 표현하기 - 선으로
sns.relplot(data = flights, x = 'year', y = 'passengers', kind = 'line')

# 선 - 평균, 바 - 신뢰구간 95%
```

```python
sns.relplot(data = flights, x = 'month', y = 'passengers', hue = 'year', kind = 'line')
```

```python
sns.relplot(data = flights, x = 'month', y = 'passengers', col = 'year',col_wrap = 4, kind = 'line')

# 꼭 유용하지는 않음, 할 수는 있다.
```

```python
# x : 범주형, y : 연속형 변수

sns.catplot(data = data, x = 'species', y = 'body_mass_g',kind = 'swarm') # kind = swarm얼만큼 많은 데이터가 몰려있는지 알기 쉽게

# 몸무게 분포
```

```python
sns.catplot(data = data, x = 'species', y = 'body_mass_g',kind = 'box', hue='sex' )
```

```python
# boxplot 눞히기

sns.catplot(data = data, y = 'species', x = 'body_mass_g',kind = 'box', hue='sex', aspect = 2 ) # aspect : 세로 1 -> 가로 2배로 올리기
```

```python
mpg = sns.load_dataset("mpg") # 자동차 데이터
```

```python
sns.catplot(data = mpg, x = 'origin', y = 'mpg', kind = 'bar') # catplot default = scatterplot, 검은 막대 : 신뢰구간 95%
```

```python
sns.displot(data = mpg,  x = 'weight', kind = 'hist', kde = True, col = 'origin', bins = 30) # displot : 분포로 표현해주는 plot

# 대형차, 소형차
```

```python
mpg.head()
```

```python
corr = mpg.corr()
```

```python
sns.heatmap(corr,cmap = 'vlag')
```

```python
pv = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(pv, annot = True, fmt = 'd') # passenger가 점점 많아짐, 많아졌다가 줄어듬 annot : 숫자를 넣고 싶을 때 , fmt = 'd'(정수)
```
