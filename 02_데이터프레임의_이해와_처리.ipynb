### Indexing / Slicing

|Methods|Description|
|--|--|
|`df[value]`|하나의 컬럼 또는 여러 컬럼 선택|
|`df.loc[value]`|value값(Index값 기준)으로 행의 부분집합을 선택|
|`df.loc[:, value]`|value값으로 열의 부분집합을 선택|
|`df.loc[value1, value2]`|value값으로 행과 열의 부분집합을 선택|
|`df.iloc[where]`|where 값(순서값 기준)으로 행의 부분집합 선택|
|`df.iloc[:, where]`|where 값으로 열의 부분집합 선택|
|`df.iloc[where1, where2]`|where 값으로 행과 열의 부분집합을 선택|

## 데이터 기본연산

```python
tmp = movie.select_dtypes('number') # int, float 모두 선택 가능
```

```python
(tmp / 100).round(2) # 마치 100으로 채워져 있는 행렬로 계산시켜줌, broadcasting, 둘째자리 까지 반올림
```

```python
tmp_mean = tmp.mean(axis = 0) # 세로 = 0, 가로 기준 = 1
```

```python
tmp.div(tmp_mean).round(2) # 각 열에 mean이 적용됨, 반올림 소수점 둘째자리
```

```python
tmp['num_critic_for_reviews'].quantile([.25,.5,.75]) # 1,2,3 사분위수를 알려줘
```
