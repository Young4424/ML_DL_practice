## 데이터 및 메타데이터 처리

### Json 파일 로드 함수, 데이터셋을 페르소나 별로 필터링하기

```python
import json
import random
import torch
from torch.utils.data import Dataset, DataLoader  # Dataset과 DataLoader 임포트
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
```

```python
# JSON 파일 로드 및 페르소나 필터링 함수
def load_and_filter_by_persona(file_path, age_id='A02', gender_id='G02'):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 페르소나 기준으로 필터링 (연령 A02: 청년, 성별 G02: 남성)
    filtered_data = [entry for entry in data if (
        entry['profile']['persona']['persona-id'].startswith(age_id) and
        entry['profile']['persona']['persona-id'].find(gender_id) != -1
    )]

    return filtered_data

# 샘플 크기를 조정하는 함수 (100개 이하로 사용)
def sample_filtered_data(filtered_data, sample_size=100):
    return random.sample(filtered_data, min(sample_size, len(filtered_data)))
```

```python
# Training, Validation 데이터 로드
filtered_train_data = load_and_filter_by_persona('/content/emotion_conversation/Training/labeling/emotion_Training.json',
                                        age_id='A02',gender_id='G02')
filtered_valid_data = load_and_filter_by_persona('/content/emotion_conversation/Validation/labeling/emotion_Validation.json',
                              age_id='A02',gender_id='G02')
```

```python
# 데이터에서 대화(HS와 SS)를 추출하여 리스트로 변환
def extract_conversations(data):
    conversations = []
    for entry in data:
        content = entry['talk']['content']
        for i in range(1, 4):  # HS01, HS02, HS03, SS01, SS02, SS03를 대상으로 함
            hs_key = f'HS{i:02}'
            ss_key = f'SS{i:02}'
            if content.get(hs_key) and content.get(ss_key):
                conversations.append((content[hs_key], content[ss_key]))
    return conversations

train_conversations = extract_conversations(filtered_train_data)
valid_conversations = extract_conversations(filtered_valid_data)
```

```python
class ConversationDataset(Dataset):
    def __init__(self, conversations, tokenizer, max_length=512):
        self.conversations = conversations
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.conversations)

    def __getitem__(self, idx):
        hs, ss = self.conversations[idx]
        input_text = f"User: {hs} System:"
        label_text = f" {ss}"

        # 각 배치마다 토큰화
        inputs = self.tokenizer(input_text, return_tensors="pt", padding='max_length', truncation=True, max_length=self.max_length)
        labels = self.tokenizer(label_text, return_tensors="pt", padding='max_length', truncation=True, max_length=self.max_length)

        # 패딩 토큰은 손실 계산에서 제외
        labels['input_ids'][labels['input_ids'] == self.tokenizer.pad_token_id] = -100

        return {
            'input_ids': inputs['input_ids'].squeeze(),
            'attention_mask': inputs['attention_mask'].squeeze(),
            'labels': labels['input_ids'].squeeze()
        }

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(base_model)

# 학습 및 검증 데이터셋 생성
train_dataset = ConversationDataset(train_conversations, tokenizer)
valid_dataset = ConversationDataset(valid_conversations, tokenizer)
```

```python
from transformers import AdamW, get_linear_schedule_with_warmup
# 모델 로드
model = AutoModelForCausalLM.from_pretrained(base_model)

# Eager Attention: 더 명확하고 디버깅하기 쉬운 구현을 제공하며, 학습 과정에서의 성능이 SDPA보다 더 안정적인 경우가 많다고 함.
```

```python
# TrainingArguments에서 deepspeed 옵션 추가
training_args = TrainingArguments(
    output_dir='./results',                     # 출력 디렉토리
    per_device_train_batch_size=1,              # 배치 크기
    per_device_eval_batch_size=1,               # 평가 배치 크기
    num_train_epochs=3,                         # 학습 에포크 수
    warmup_steps=500,                           # 워밍업 스텝 수
    weight_decay=0.01,                          # 가중치 감소율
    logging_dir='./logs',                       # 로그 디렉토리
    logging_steps=10,                           # 로그 빈도
    deepspeed='/content/deepspeed_config.json',        # Deepspeed 설정 파일 경로
    fp16=True                                   # 16-bit 부동소수점 연산 활성화
)
```

```python
# Trainer 설정
trainer = Trainer(
    model=model,                              # 모델
    args=training_args,                       # 학습 설정
    train_dataset=train_dataset,              # 학습 데이터셋
    eval_dataset=valid_dataset                # 검증 데이터셋
)

# 학습 시작
trainer.train()
```

## 토크나이저와 모델 로드

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# 학습된 모델과 토크나이저 로드
model = AutoModelForCausalLM.from_pretrained('./results')  # 학습된 모델 경로
tokenizer = AutoTokenizer.from_pretrained('./results')    # 학습된 토크나이저 경로

# 모델을 평가 모드로 설정
model.eval()

# 사용자가 일기를 입력하면 이에 대한 반응을 생성하는 함수
def generate_response(user_input, model, tokenizer, max_length=100):
    # 사용자 입력을 토큰화
    input_text = f"User: {user_input} System:"
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # 모델을 사용하여 답변 생성
    with torch.no_grad():
        output = model.generate(
            input_ids=input_ids,
            max_length=max_length,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            early_stopping=True
        )

    # 생성된 토큰을 텍스트로 변환
    response = tokenizer.decode(output[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

    return response

# 사용자의 일기를 입력받아 모델이 답변을 생성
user_input = input("일기를 입력하세요: ")
response = generate_response(user_input, model, tokenizer)

print(f"AI의 반응 및 조언: {response}")
```
